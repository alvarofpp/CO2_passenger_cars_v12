# Comandos usados
- **.dropna()** - Tira as linhas que contém cédulas NaN;
- **.drop_duplicates** - Remove os valores duplicatos.

```python
from bokeh.plotting import figure # Import figure from bokeh.plotting
from bokeh.io import output_notebook, show # Import output_notebook and show from bokeh.io
import pandas as pd # Import pandas as pd
import numpy as np # Import numpy as np
from bokeh.charts import Bar # Gráfico
from bokeh.layouts import row

data = pd.read_csv("CO2_passenger_cars.csv", encoding = 'latin2', low_memory=False)
```

```python
data_mp = data[['MP', 'ep (KW)', 'e (g/km)']].dropna()

# MP
dados_solicitados = ['ep (KW)', 'e (g/km)']
data_volkswagen = data_mp[data_mp['MP'] == 'VW GROUP PC'][dados_solicitados]
data_toyota = data_mp[data_mp['MP'] == 'TOYOTA-DAHAITSU GROUP'][dados_solicitados]
data_daimler = data_mp[data_mp['MP'] == 'DAIMLER AG'][dados_solicitados]
data_bmw = data_mp[data_mp['MP'] == 'BMW GROUP'][dados_solicitados]
data_hyundai = data_mp[data_mp['MP'] == 'HYUNDAI'][dados_solicitados]
data_fca = data_mp[data_mp['MP'] == 'FCA ITALY SPA'][dados_solicitados]
data_ford = data_mp[data_mp['MP'] == 'FORD-WERKE GMBH'][dados_solicitados]
data_mitsubishi = data_mp[data_mp['MP'] == 'MITSUBISHI MOTORS'][dados_solicitados]
data_renault = data_mp[data_mp['MP'] == 'RENAULT'][dados_solicitados]
data_general = data_mp[data_mp['MP'] == 'GENERAL MOTORS'][dados_solicitados]
data_suzuki = data_mp[data_mp['MP'] == 'SUZUKI POOL'][dados_solicitados]
data_kia = data_mp[data_mp['MP'] == 'KIA'][dados_solicitados]
data_tata = data_mp[data_mp['MP'] == 'TATA MOTORS LTD'][dados_solicitados]
data_honda = data_mp[data_mp['MP'] == 'HONDA MOTOR EUROPE LTD'][dados_solicitados]
```

```python
data_ft = data[['Ft', 'ep (KW)', 'e (g/km)']].dropna()
data_diesel = data_ft[data_ft['Ft'] == 'DIESEL'][['ep (KW)', 'e (g/km)']]
data_petrol = data_ft[data_ft['Ft'] == 'PETROL'][['ep (KW)', 'e (g/km)']]
data_electric = data_ft[data_ft['Ft'] == 'ELECTRIC'][['ep (KW)', 'e (g/km)']]
data_biomethane = data_ft[data_ft['Ft'] == 'NG-BIOMETHANE'][['ep (KW)', 'e (g/km)']]

p = figure(x_axis_label='Electric energy consumption (Wh/hm)', y_axis_label='Specific CO2 Emissions (g/km)')

# ForestGreen
p.circle(data_diesel['ep (KW)'], data_diesel['e (g/km)'], color='#228B22', size=4, alpha=0.7, 
         legend='Diesel')
# SlateGray
p.circle(data_petrol['ep (KW)'], data_petrol['e (g/km)'], color='#708090', size=4, alpha=0.7, 
        legend='Petrol')
# Goldenrod
p.square(data_electric['ep (KW)'], data_electric['e (g/km)'], color='#DAA520', size=5, alpha=1.0,
        legend='Electric')
# GreenYellow
p.square(data_biomethane['ep (KW)'], data_biomethane['e (g/km)'], color='#ADFF2F', size=5, alpha=1.0,
        legend='Biomethane')

# Assign the legend to the bottom left: p.legend.location
p.legend.location = 'top_left'
# Fill the legend background with the color 'lightgray': p.legend.background_fill_color
p.legend.background_fill_color='lightgray'

output_notebook()
show(p)
```

```python
data_m = data[['m (kg)', 'e (g/km)', 'ep (KW)']].dropna()

p1 = figure(x_axis_label='Mass (kg)', y_axis_label='Specific CO2 Emissions (g/km)')
p1.circle(data_m['m (kg)'], data_m['e (g/km)'], color='blue', size=5, alpha=0.8)

p2 = figure(x_axis_label='Mass (kg)', y_axis_label='Electric energy consumption (Wh/hm)')
p2.circle(data_m['m (kg)'], data_m['ep (KW)'], color='red', size=5, alpha=0.8)

layout = row(p1,p2,sizing_mode='scale_width')

output_notebook()
show(layout)
```

```python
data_er = data[['e (g/km)', 'Er (g/km)', 'ep (KW)']].dropna()

p1 = figure(x_axis_label='Engine power (KW)', y_axis_label='Specific CO2 Emissions (g/km)')
p1.circle(data_er['ep (KW)'], data_er['e (g/km)'], color='blue', size=5, alpha=0.8)

p2 = figure(x_axis_label='Engine power (KW)', 
            y_axis_label='Emissions reduction through innovative technologies (g/km)')
p2.circle(data_er['ep (KW)'], data_er['Er (g/km)'], color='red', size=5, alpha=0.8)

layout = row(p1,p2,sizing_mode='scale_width')

output_notebook()
show(layout)
```
